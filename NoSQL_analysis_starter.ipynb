### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20 }

# Display the number of documents in the result
results = establishments.find(query).explain().get("executionStats").get("nReturned")

print(f'Number of establishments with a hygiene score of 20 : {results}')
```

```python
# Use count_documents to display the number of documents in the result
result = establishments.count_documents(query)
print(f'Number of establishments with a hygiene score of 20 : {result}')
```

```python
# Display the first document in the results using pprint

results = list(establishments.find(query))

pprint(results[0])
```

```python
#extracting latitude, longitude and hygiene score in 3 separate functions

def get_latitude(geocode):
    return geocode.get('latitude')

def get_longitude(geocode):
    return geocode.get('longitude')

def get_hygiene_score(row):
    hygiene_score = row.get('Hygiene')
    return hygiene_score
```

```python
# Convert the result to a Pandas DataFrame

hygiene20_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Number of rows in the dataframe for hygiene score equal to 20 : {len(hygiene20_df)}")

#dataframe cleaning
hygiene20_df['latitude'] = hygiene20_df['geocode'].apply(get_latitude)
hygiene20_df['longitude'] = hygiene20_df['geocode'].apply(get_longitude)
hygiene20_df['Hygiene'] = hygiene20_df['scores'].apply(get_hygiene_score)

#filtering the columns
hygiene20_clean_df = hygiene20_df[['BusinessName','RatingValue','Hygiene','longitude','latitude']]


# Display the first 10 rows of the DataFrame
hygiene20_clean_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

#get the coordinates of Penang Flavors using its _id which was determined while inserting record into db:
query = {"_id":objInstance}
fields = {'geocode.longitude' : 1, 'geocode.latitude' : 1}
results = establishments.find(query,fields)

for result in results:
    lat=result.get('geocode').get('latitude')
    lat_from = result.get('geocode').get('latitude') - 0.01
    lat_to = result.get('geocode').get('latitude') + 0.01
    lon=result.get('geocode').get('longitude')
    lon_from=result.get('geocode').get('longitude') - 0.01
    lon_to=result.get('geocode').get('longitude') + 0.01
    
    match_query = {
                    '$match': 
                            {
                               "_id": { '$ne' : objInstance },
                              'geocode.longitude': {'$gte': lon_from, '$lt': lon_to},
                              'geocode.latitude': {'$gte': lat_from, '$lt': lat_to},
                              'RatingValue': 5  
                            }
                  }
    #sort by Hygiene in descending order
    sort_values = {'$sort': { 'scores.Hygiene': 1 }}         

    # Put the pipeline together
    pipeline = [match_query, sort_values]    
    
    # Run the pipeline through the aggregate method, cast the results as a list, and save the results to a variable
    pipeline_result = list(establishments.aggregate(pipeline))
    print(f'Number of restaurants with Rating value of 5 and good Hygiene score: {len(pipeline_result)}')
    limit =5

    #Print the results
    print("Top 5 establishments with Rating 5 and having a good Hygeine score : -")
    top_n = pipeline_result[0:limit] if len(pipeline_result) > limit + 1 else pipeline_result
    pprint(top_n)
```

```python
# Convert result to Pandas DataFrame: Top 5 establishments with a rating value of 5 and sorted by lowest hygiene score

top5_nearby_establishments = pd.DataFrame(top_n)
top5_nearby_establishments 
```

```python
#dataframe cleaning
top5_nearby_establishments['latitude'] = top5_nearby_establishments['geocode'].apply(get_latitude)
top5_nearby_establishments['longitude'] = top5_nearby_establishments['geocode'].apply(get_longitude)
top5_nearby_establishments['Hygiene'] = top5_nearby_establishments['scores'].apply(get_hygiene_score)

#filtering the columns
top5_nearby_est_clean = top5_nearby_establishments[['BusinessName','RatingValue','Hygiene','longitude','latitude']]


# Display the DataFrame
print("Top 5 restaurants nearby to Penang Flavors--> ")
top5_nearby_est_clean
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {
                    '$match': 
                            { 'scores.Hygiene' : 0
                            }
                  }

group_query = {'$group': { '_id': {'LocalAuthorityName': '$LocalAuthorityName'}, 'count': { '$sum': 1 }}}

#sort by Hygiene in descending order
sort_values = {'$sort': { 'count': -1 }}         

# Put the pipeline together
pipeline = [match_query,group_query, sort_values]    

# Run the pipeline through the aggregate method, cast the results as a list, and save the results to a variable
pipeline_result = list(establishments.aggregate(pipeline))

limit =10

# Print the number of documents in the result
print(f'Total Number of documents with hygiene score 0 : {len(pipeline_result)}')

print('Top 10 establishments by count: ')

# Print the first 10 results
top_n = pipeline_result[0:limit] if len(pipeline_result) > limit + 1 else pipeline_result
pprint(top_n)




```

```python
#function to fetch the localauthorityname from the _id field
def get_local_auth_name(x):
    return x.get('LocalAuthorityName')

local_auth_df = pd.DataFrame(pipeline_result)
local_auth_df['LocalAuthorityName'] = local_auth_df['_id'].apply(get_local_auth_name)

local_auth_df.head()
```

```python
#rearranging columns
local_auth_df = local_auth_df[['LocalAuthorityName','count']]
# Display the first 5 rows of the DataFrame
local_auth_df.head()
```

### Extracting all establishments having 'London' as Local Authority Name - Map plots for later

```python
#fetching all the establishments having London as their Local Authority
query = {'LocalAuthorityName' : {'$regex': 'London'}}

london_est_results = list(establishments.find(query))
```

```python
#Converting to a dataframe
london_est_df = pd.DataFrame(london_est_results)

#reading the columns
london_est_df.columns

```

```python
#normalizing json for extracting the coordinates into a dataframe
df_coordinates = pd.json_normalize(london_est_df['geocode']) 

#normalizing json for extracting the scores into a dataframe
df_scores = pd.json_normalize(london_est_df['scores'])

#creating a dataframe that contains Rating value and date along with Business name
df_rating = london_est_df[['BusinessName','RatingValue','RatingDate']]

#Merging the dataframes
df_rating = pd.merge(df_rating,df_scores,left_index=True,right_index=True,how='inner')
df_rating = pd.merge(df_rating,df_coordinates,left_index=True,right_index=True,how='inner')

#displaying the dataframe
df_rating.head()
```

```python
#writing the dataframe into a csv to be used for the map plot in London_Ratings_Map_plots.ipynb. 
df_rating.to_csv('london_ratings.csv',index=False)
```
