```python
import pandas as pd
import plotly_express as px
from datetime import datetime
```

```python
def parse_time(text):
    
    dt = datetime.strptime(text,"%Y-%m-%dT00:00:00")
    return dt
```

```python
datafile = "all_establishments_rating.csv"

establishments_df = pd.read_csv(datafile,sep=',')

establishments_df.head()
```

```python
establishments_df.dropna(inplace=True)

establishments_df['RatingDate'] = establishments_df['RatingDate'].apply(parse_time)

establishments_df.head()
```

```python
establishments_df['Rating_Rank'] = establishments_df['RatingValue'].rank(method='dense',ascending=False)

establishments_df.head(10)
```

```python
establishments_df.dtypes
```

```python
#fig = px.density_mapbox(london_df,lat='latitude', lon='longitude',z='Rating_Rank',mapbox_style='stamen-toner',zoom=10)
fig = px.scatter_mapbox(establishments_df.head(15),lat='latitude', lon='longitude',size='Rating_Rank',mapbox_style='open-street-map',zoom=6,
                        opacity=0.5,title='Establishments by Rating and Hygiene')

lats = list(establishments_df['latitude'])
lons = list(establishments_df['longitude'])
sizes = list(establishments_df['Hygiene'])


fig.add_scattermapbox(lat=lats, lon=lons,opacity=0.2,name='Hygiene',marker_size= sizes)


fig.show()
```
